```python
X=df.drop('quality', axis=1)
y=df['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_scaled= pd.DataFrame(scale.fit_transform(X),columns =X.columns)
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2, random_state=42)
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
prediction_test=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('training accuracy:', accuracy_score(y_test,prediction_test))
```

```python
pd.crosstab(y_test,prediction_test)
```

```python
new_input_data = (7.2, 0.6, 0.39, 5.8, 0.066, 15.5, 98.5, 0.9970, 3.28, 0.75, 9.8)

input_data_array = np.asarray(new_input_data)

# Reshape the input data for prediction
reshaped_input = input_data_array.reshape(1, -1)

predicted_quality = model.predict(reshaped_input)

if predicted_quality[0] == 1:
    print('This is a high-quality wine.')
else:
    print('This is a lower-quality wine.')
```
